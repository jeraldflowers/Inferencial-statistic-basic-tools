### 2. Systematic Sampling

```python
def systematic_sampling(econdata, step):
  indexes = np.arange(0, len(econdata), step=step)
  systematic_sample = econdata.iloc[indexes]
  return systematic_sample

systematic_sample = systematic_sampling(econdata, 3)
systematic_sample
```

### 3. Stratified Sampling

```python
econdata['estratificado'] = econdata ['delegacion'] + ", " + econdata['tipo']
(econdata['estratificado'].value_counts()/len(econdata)).sort_values(ascending=False)
```

```python
def data_stratified(econdata, name_col_strat, values_strat, prop_strat, random_state=None):
  df_strat = pd.DataFrame(columns=econdata.columns)

  pos = -1
  for i in range(len(values_strat)):
    pos += 1
    if pos == len(values_strat) - 1:
      ratio_len = len(econdata) - len(df_strat)
    else:
      ratio_len = int(len(econdata) * prop_strat[i])


    df_filtered = econdata[econdata[name_col_strat] == values_strat[i]]
    df_temp = df_filtered.sample(replace=True, n=ratio_len, random_state=random_state)

    df_strat = pd.concat([df_strat, df_temp])

  return df_strat
```

```python
values_strat = ['Cuautémoc, Hotel', 'Cuautémoc, Museo', 'Venustiano Carranza, Hotel' ,'Cuauhtémoc, Mercado' ,'Venustiano Carranza, Mercado']
prop_strat = [0.5, 0.2, 0.1, 0.1, 0.1]
df_strat = data_stratified(econdata, 'estratificado', values_strat, prop_strat, random_state = 42)
df_strat
```
